```python
''' Mathematics Forumula :  σ(x)=1/ 1+e−x '''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)


x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.show()

print("Sigmoid of 0:", sigmoid(0))
print("Derivative of 0:", sigmoid_derivative(0))
```
