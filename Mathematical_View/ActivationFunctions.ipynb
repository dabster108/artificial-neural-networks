```python
''' Sigmoid Activation Function  Mathematics Forumula :  σ(x)=1/ 1+e−x '''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)


x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.show()

print("Sigmoid of 0:", sigmoid(0))
print("Derivative of 0:", sigmoid_derivative(0))
```

```python
import numpy as np 
# Tanh function
def tanh(x):
    return np.tanh(x)

# Derivative
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Plot
y = tanh(x)

plt.plot(x, y)
plt.title("Tanh Function")
plt.show()

print("Tanh of 0:", tanh(0))
print("Derivative of 0:", tanh_derivative(0))
```

```python
''' ReLU(Rectified Linear Unit) Activation Function  Mathematics Forumula :  ReLU(x)=max(0,x) '''
```

```python
import numpy as np 
# ReLU function
def relu(x):
    return np.maximum(0, x)

# Derivative
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Plot
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.show()

print("ReLU of -3:", relu(-3))
print("Derivative of -3:", relu_derivative(-3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Input range
x = np.linspace(-10, 10, 400)

# Compute values
sigmoid_y = sigmoid(x)
tanh_y = tanh(x)
relu_y = relu(x)

# Create plot
plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid_y, label="Sigmoid", color="blue")
plt.plot(x, tanh_y, label="Tanh", color="green")
plt.plot(x, relu_y, label="ReLU", color="red")

# Styling
plt.axhline(0, linestyle="--", linewidth=0.8)
plt.axvline(0, linestyle="--", linewidth=0.8)
plt.title("Activation Functions Comparison")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```
