```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1,2],
    [2,3],
    [3,3],
    [5,6]
])

y = np.array([
    [0],
    [0],
    [1],
    [1]
])
```

```python
np.random.seed(0)

W1 = np.random.randn(2,2)
b1 = np.zeros((1,2))

W2 = np.random.randn(2,1)
b2 = np.zeros((1,1))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
Z1 = np.dot(X,W1) + b1
A1 = relu(Z1)

Z2 = np.dot(A1,W2) + b2
A2 = sigmoid(Z2)

print(A2)
```

```python
loss = -np.mean(y*np.log(A2) + (1-y)*np.log(1-A2))
print(loss)
```
